```python
import pandas
import numpy
import matplotlib.pyplot
import sklearn.manifold
```

```python
employees = pandas.read_csv('employee_info.csv')
employees
```

What we want to do is create a file in which we can easily determine if an employee was hired or fired, if an employee changed roles or if an employee changed supervisors. 

```python
group = employees.groupby('employee_name')
```

```python
role = group.role.aggregate(' '.join)
months = group.month.aggregate(' '.join)
```

```python
role = role.apply(lambda t: [w for w in t.split()]  )
months = months.apply(lambda b:[w for w in b.split()])
role = role.apply(lambda bag: pandas.Series(bag).value_counts())
months = months.apply(lambda bag: pandas.Series(bag).value_counts())


role
```

```python
months
```

```python
calendar = ['June', 'July', 'August', 'September', 'October', 'November']

stoppedWorking = pandas.DataFrame()

for i in range(0,5):
    str1 = '%d/1/17' % (i+6)
    str2 = '%d/1/17' % (i+7)

    employeeChangeJune = months[months[str1] != 1.0]
    months = months[months[str1] == 1.0]

    stoppedWorking1 = pandas.DataFrame()
    stoppedWorking1['employee'] = employeeChangeJune.index.values
    stoppedWorking1['month'] = calendar[i]

    employeeChangeJuly = months[months[str2] != 1.0]
    months = months[months[str2] == 1.0]

    stoppedWorking2 = pandas.DataFrame()
    stoppedWorking2['employee'] = employeeChangeJuly.index.values
    stoppedWorking2['month'] = calendar[i]

    stoppedWorking1 = stoppedWorking1.append(stoppedWorking2)
    stoppedWorking = stoppedWorking.append(stoppedWorking1)
    
    i = i + 1
```

```python
notWorkingRoles = []
supervisors = []
for employee in stoppedWorking.employee:
    person = employees[employees['employee_name'] == employee]
    supervisor = person.supervisor.values
    supervisor = supervisor[0]
    role2 = person.role.values
    role2 = role2[0]
    supervisors.append(supervisor)
    notWorkingRoles.append(role2)

stoppedWorking['role'] = notWorkingRoles
stoppedWorking['supervisor'] = supervisors
stoppedWorking
```

```python
stoppedWorking.to_csv('fired.csv')
```
